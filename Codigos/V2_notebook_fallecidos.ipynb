# Estadisticas 

```python
df = consolidado_prueba.copy()
```

```python
fecha_hoy = date.today()
estadisticas_por_problema = df.groupby('Problema Oncologico').apply(lambda group: pd.Series({
    'Fecha Medición': fecha_hoy,
    'Pacientes en Etapa no Alta': group[group['Etapa']!='Alta'].shape[0],
    'Pacientes en Etapa Alta y Estado Fallecido': group[(group['Etapa'] =='Alta') & (group['Estado'] =='Fallecido')].shape[0],
    'Pacientes en Etapa Alta pero Estado No Fallecido': group[(group['Etapa'] =='Alta') & (group['Estado'] !='Fallecido')].shape[0]
}))
```

```python
estadisticas_por_problema.reset_index(inplace= True)
```

```python
estadisticas_por_problema
```

```python
archivos_est = glob.glob(os.path.join(dir_actual, '..','Power BI','*Estadisticas_por_problema*'))

archivos_ordenados_est = sorted(archivos_est, key=os.path.getmtime, reverse = True)

rutas_archivos = archivos_ordenados_est[:1]
```

```python
rutas_archivos
```

```python
df_existente = pd.read_excel(rutas_archivos[0])
df_existente['Fecha Medición'] = pd.to_datetime(df_existente['Fecha Medición'])
```

```python
df_existente
```

```python
df_existente['Problema Oncologico'].value_counts()
```

```python
df_existente['Problema Oncologico'] = df_existente['Problema Oncologico'].replace('Cáncer Tiroide','Cáncer Tiroides')
```

```python
import pandas
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows
```

```python
# Día actual
today = date.today()

# Fecha actual
now = datetime.now()

# Formatear día con ceros a la izquierda si es necesario
day_formatted = "{:02d}".format(today.day)
# Formatear mes con ceros a la izquierda si es necesario
month_formatted = "{:02d}".format(today.month)
```

```python
wb = Workbook()
ws = wb.active
df_final = pd.concat([df_existente,estadisticas_por_problema], ignore_index=True)

for r in dataframe_to_rows(df_final, index= False, header = True):
    ws.append(r)

for cell in ws['B'][1:]:
    cell.number_format = 'DD-MM-YYYY'
nombre_archivo = f'Estadisticas_por_problema_{today.year}{month_formatted}{day_formatted}.xlsx'
wb.save(os.path.join(dir_actual, '..','Power BI',f'{nombre_archivo}'))
```
